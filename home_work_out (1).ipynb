Write a function to calculate the sum of digits in a number entered by user

```python
def sum_of_digit():
  num=input('enter the number:')
  sum=0
  for i in num:
    sum+=int(i)
  print(sum)
sum_of_digit()
```

Create a list "mylist" that contains integer numbers, float numbers, and strings.

Now write a function that creates a dictionary with each data types and its counts.

if the list is mylist = [1, 7, 15, 25, 1.5, 55.8, "Arun", "Akhil", "Akshay"],

then the dictionary will have {"integer numbers": 4, "floating numbers":2, "Strings": 3}

```python
mylist=[7,1,15,25,1.5,55.8,'Arun','Akhil','Akshay']

def count_values(lst):
  type_count={'integer':0,'floating number':0,'strings':0}
  for i in lst:
    if type(i) == int:
      type_count['integer']+=1
    elif type(i)== float:
      type_count['floating number']+=1
    else:
      type_count['strings']+=1
  print(type_count)
count_values(mylist)




```

Create a function that takes a string and returns a dictionary with counts of vowels and consonants.

```python
def vowels():
  string=input('enter the string:')
  result={'vowel':0,'consonent':0}
  for i in string:
      if i in 'AEIOUaeiou':
        result['vowel']+=1
      else:
        result['consonent']+=1
  print(result)
vowels()
```

Write a function that checks if a number is a strong number

(sum of factorials of digits equals the number, e.g., 145 = 1! + 4! + 5!).

```python
import math
def strong_number():
  num=input('enter the number:')
  sum=0
  for i in num:
    sum+=math.factorial(int(i))
  print('strong number of',num,'is :',sum)
strong_number()


```
