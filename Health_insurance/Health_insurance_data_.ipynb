```python
#listing the unique values of each column
for i in df.select_dtypes(include=['object','float','integer']).columns:
  uniq_vals=df[i].unique()
  print(f"columns '{i}' : ({len(uniq_vals)})")
  print(uniq_vals)
  print('*'*30)
```

```python
#proportion of missingness
for i in df[['Health Indicator','Holding_Policy_Duration','Holding_Policy_Type']]:
  missing_pro=df[i].isna().mean()
  print(f"column {i}:{missing_pro}")
```

```python
df.T
```

```python
#droping id column
df.drop(['ID'],axis=1,inplace=True)
```

```python
df
```

```python
#filling the missing values
df['Health Indicator']=df['Health Indicator'].fillna(df['Health Indicator'].mode()[0])
```

```python
df['Holding_Policy_Duration']=df['Holding_Policy_Duration'].fillna(df['Holding_Policy_Duration'].mode()[0])
```

```python
df['Holding_Policy_Type']=df['Holding_Policy_Type'].fillna(df['Holding_Policy_Type'].mode()[0])
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
#Explorary Data Analysis And Feature Engineering
df['Response']=df['Response'].replace({'No':0,'Yes':1})
```

```python
#ploting
sns.countplot(x='Response',data=df)
plt.title('Response for Health Insurance ')
plt.title('Response')
plt.show()
```

```python
df.Response.value_counts().plot.pie(autopct='%.1f%%')
plt.show()
```

```python
#policy type
sns.countplot(x='Holding_Policy_Type',data=df)
plt.title('Policy Type of Data')
plt.show()
print(df.Holding_Policy_Type.value_counts(normalize=True)*100)
```

```python
df.columns
```

```python
df['Reco_Insurance_Type']=df['Reco_Insurance_Type'].replace({'Individual':1,'Joint':2})
```

```python
sns.countplot(x='Reco_Insurance_Type',data=df)
plt.title('Insurance Type ')
plt.show()
print(df.Reco_Insurance_Type.value_counts(normalize=True)*100)
```

```python
sns.countplot(x='Holding_Policy_Duration',data=df)
plt.title(' Policy Duration')
plt.show()
print(df.Holding_Policy_Duration.value_counts(normalize=True)*100)
```

```python
df['Holding_Policy_Duration']=df['Holding_Policy_Duration'].replace(10,10)
df['Holding_Policy_Duration']=df['Holding_Policy_Duration'].replace(11,10)
df['Holding_Policy_Duration']=df['Holding_Policy_Duration'].replace(12,10)
df['Holding_Policy_Duration']=df['Holding_Policy_Duration'].replace(13,10)
df['Holding_Policy_Duration']=df['Holding_Policy_Duration'].replace(14,10)
```

```python
#Heat map
columns=df.columns
cols=df[columns].select_dtypes(include=['number']).columns.tolist()

corr_matrix=df[cols].corr()

plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix,annot=True,fmt='.2f',cmap='coolwarm',center=0)
plt.title('Heat Map')
```
